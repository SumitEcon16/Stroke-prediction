```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('E:\Rough\healthcare-dataset-stroke-data.csv')
```

```python
df.head(5)
```

```python
sex = list(df['gender'])
male = sex.count('Male')
female = sex.count('Female')
others = len(sex) - male - female

label = ['Male','Female','Others']
data = [male,female,others]
plt.pie(data, labels= label)
plt.show()
```

```python
gender = []
for i in sex:
    if i == 'Male':
        gender.append(1)
    elif i == 'Female':
        gender.append(0)
    else:
        gender.append(2)
df.drop('gender', axis = 1, inplace=True)
df['gender'] = gender
df
```

```python
work = list(df['work_type'])
label = []
data = []
wrk = pd.Series(work).value_counts()
for x,y in wrk.items():
    label.append(x)
    data.append(y)
print(data, label)
plt.pie(data, labels= label)
plt.show()
```

```python
smoke = list(df['smoking_status'])
smk = pd.Series(smoke).value_counts()
label = []
data = []
for x,y in smk.items():
    label.append(x)
    data.append(y)
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd',
          '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
ll= len(label)
plt.bar(label, data, width = 0.4, color=colors[0:ll])
```
