In this project, our objective is to study the data and observe it's features, and then use the data to create a model which predicts the chance of heart stroke. We start by importing the libraries we require.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('E:\Rough\d-healthcare-dataset-stroke-data.csv')
```

```python
stroke = df.drop(columns='id', axis=1)
stroke.head(15)
```

```python
stroke.isnull().sum()
```

```python
bmi = stroke['bmi']

bmi_mean = bmi.mean()
stroke = stroke.fillna(bmi_mean)

stroke
```

```python
x = stroke.drop(columns='stroke', axis=1)
y = stroke['stroke']
```

```python
x.isnull().sum()
```

```python
bmi = x['bmi']

bmi_mean = bmi.mean()
x = x.fillna(bmi_mean)
x.isnull().sum()
```

```python
#now we split the data into train and test sets

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
#now we train our model

model = LogisticRegression()
model.fit(x_train, y_train)
```

```python
# model evaluation with accuracy score

x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)

print(training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
test_data_accuracy = accuracy_score(x_test_prediction, y_test)

print(test_data_accuracy)
```
